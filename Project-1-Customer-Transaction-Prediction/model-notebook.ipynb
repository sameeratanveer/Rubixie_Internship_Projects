### Detecting and Handling Outliers.

```python
from sklearn.ensemble import IsolationForest
from scipy.stats import mstats

outlier_free_df = df 

# Initialize the Isolation Forest model
iso_forest = IsolationForest(contamination=0.05)  
outlier_labels = iso_forest.fit_predict(outlier_free_df)

# Identify outliers
outlier_free_df['Outlier'] = outlier_labels
outliers = outlier_free_df[outlier_free_df['Outlier'] == -1]  # Outliers will be labeled as -1

# percentage of outliers
outlier_percentage = len(outliers) / len(outlier_free_df) * 100

# Determine action based on the percentage of outliers
if outlier_percentage < 10:
    # Drop the outliers
    df_cleaned = outlier_free_df[outlier_free_df['Outlier'] != -1]
    print(f"Dropped outliers. Remaining data shape: {df_cleaned.shape}")
else:
    # Imputation or Winsorization
    df_cleaned = mstats.winsorize(outlier_free_df, limits=[0.05, 0.05])  # Clipping top and bottom 5%
    print("Applied Winsorization to the data.")
```

```python
df_new = df_cleaned
df_cleaned.loc[:, 'target'] = df_cleaned['target'].astype(int)
print("Number of missing values in target variable:", df_cleaned['target'].isnull().sum())
print("Number of infinite values in target variable:", np.isinf(df_cleaned['target']).sum())
print(f"Mode : {df_cleaned['target'].mode()[0]}")
df_cleaned.loc[:, 'target'] = df_cleaned['target'].fillna(value=df_cleaned['target'].mode()[0])
print("Number of missing values in target variable:", df_cleaned['target'].isnull().sum())
print("Number of infinite values in target variable:", np.isinf(df_cleaned['target']).sum())
```

```python
df_cleaned.isnull().sum()
df_cleaned = df_cleaned.dropna()
df_cleaned.isnull().sum()
```

### Splitting the data :
